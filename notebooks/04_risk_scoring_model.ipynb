```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("../data/processed/nepal_segmented.csv")

features = ['age','BMI','lifestyle_risk_score','ses_risk_score']
X = df[features]
y = df['diabetes_flag']

model = LogisticRegression()
model.fit(X, y)

df['diabetes_risk_score'] = model.predict_proba(X)[:,1]

def risk_level(score):
    if score < 0.3: return 'Low'
    elif score < 0.6: return 'Medium'
    else: return 'High'

df['diabetes_risk_level'] = df['diabetes_risk_score'].apply(risk_level)
df.to_csv("../data/processed/nepal_scored.csv", index=False)
df.head()
```
