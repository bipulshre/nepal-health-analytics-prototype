```python
import pandas as pd
```

```python
import numpy as np
```

```python
def generate_dummy_data(num_records = 10, random_seed= 42):
    rng = np.random.default_rng(random_seed)
    data = pd.DataFrame({
        'person_id': np.arange(1, num_records + 1),
        'age': rng.integers(18, 70, num_records),
        'gender': rng.choice(['Male','Female'], num_records, p=[0.5,0.5]),
        'province': rng.integers(1, 8, num_records),
        'urban_rural': rng.choice([0,1], num_records, p=[0.6,0.4]),
        'education_level': rng.choice(['None','Primary','Secondary','Higher'], num_records),
        'wealth_index': rng.choice(['Low','Medium','High'], num_records),
        'smoker': rng.choice([0,1], num_records, p=[0.75,0.25]),
        'alcohol_use': rng.choice([0,1], num_records, p=[0.7,0.3]),
        'physical_activity': rng.choice([0,1], num_records, p=[0.4,0.6]),
        'BMI': rng.integers(18, 35, num_records),
        'systolic_bp': rng.integers(100, 160, num_records),
        'diastolic_bp': rng.integers(60, 100, num_records),
    })
        # Logical health rules
    data['diabetes_flag'] = ((data['BMI'] > 27) & (data['age'] > 40)).astype(int)
    data['hypertension_flag'] = (data['systolic_bp'] > 140).astype(int)

    return data
```

```python
df = generate_dummy_data(50)
```
