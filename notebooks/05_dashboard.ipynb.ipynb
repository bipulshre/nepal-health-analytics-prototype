```python
import plotly.express as px
import pandas as pd

df = pd.read_csv("../data/processed/nepal_scored.csv")

# Average diabetes risk per segment
kpi_segment = df.groupby('segment_name')['diabetes_risk_score'].mean().reset_index()

fig = px.bar(
    kpi_segment,
    x='segment_name',
    y='diabetes_risk_score',
    title="Average Diabetes Risk by Population Segment in Nepal",
    labels={'segment_name':'Population Segment','diabetes_risk_score':'Average Diabetes Risk'}
)

fig.update_layout(yaxis_tickformat=".0%", bargap=0.3)
fig.show()
```
