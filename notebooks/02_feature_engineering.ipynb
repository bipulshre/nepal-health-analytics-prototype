```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("../data/dummy/nepal_dummy_population.csv")
```

```python
def bmi_category(bmi):
    if bmi < 18.5: return 'Underweight'
    elif bmi < 25: return 'Normal'
    elif bmi < 30: return 'Overweight'
    else: return 'Obese'
```

```python
df['BMI_category'] = df['BMI'].apply(bmi_category)
```

```python
df['lifestyle_risk_score'] = df['smoker'] + df['alcohol_use'] + (1 - df['physical_activity'])
```

```python
df['ses_risk_score'] = df['urban_rural'].apply(lambda x: 1 if x==0 else 0) +
                       df['education_level'].apply(lambda x: 1 if x in ['None','Primary'] else 0) +
                       df['wealth_index'].apply(lambda x: 1 if x=='Low' else 0)
```

```python
df.to_csv("../data/processed/nepal_features.csv", index=False)
```
